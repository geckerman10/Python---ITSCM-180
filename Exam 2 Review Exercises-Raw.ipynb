Through the following three exercises we are going to replicate the popular Wordle game in python. The original game website is: https://www.nytimes.com/games/wordle/index.html



## Exercise 1

Download the text file named **wordlist.txt** from the following link: https://bit.ly/3LnPMVt. The file contains 10,000 English words, each in a separate line.
Write a program that reads through the text file and extracts the words exactly 5 characters long (excluding the newline character at the end of each line). Write all such words to a new output txt file and call it **words_5.txt**.


## Exercise 2

Write a program that chooses a random 5-character word from the **words_5.txt** file created in the previous exercise and considers that as the answer to the Wordle game (store it in a variable called *answer*).



## Exercise 3

Now let the user to have at most 6 guesses (each a 5-char word). For each guess, provide appropriate hint using the words *Green*, *Gray*, or *Yellow* for each character (just like in the Wordle game). If at any point the correct answer was guessed, finish the game. Otherwise continue until the last guess.
